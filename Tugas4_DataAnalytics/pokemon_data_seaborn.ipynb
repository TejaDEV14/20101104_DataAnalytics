Visualizing Pok√©mon Stats with Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pkmn = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Pokemon.csv')

pkmn.head()

pkmn = pkmn.drop(['Generation', 'Legendary'],1)

sns.jointplot(x="HP", y="Attack", data=pkmn);
```
